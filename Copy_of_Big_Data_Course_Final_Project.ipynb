```python
# calculate kmer of kmers for the dataframe
# adaptation of the MathFeature package

import sys
import warnings
warnings.filterwarnings("ignore")
import numpy as np
import collections
import argparse
from Bio import SeqIO
from itertools import product
import math


def perms(ksize, characters):
    global listTotal, characteres
    listTotal = []
    permsList = [''.join(str(i) for i in x) for x in product(characters, repeat=ksize)]
    for perm in permsList:
        listTotal.append(perm)
    return


def chunks(seq, win, step):
    seqlen = len(seq)
    for i in range(0,seqlen,step):
        j = seqlen if i+win>seqlen else i+win
        yield seq[i:j]
        if j==seqlen: break
    return


def chunksTwo(seq, win):
    seqlen = len(seq)
    for i in range(seqlen):
        j = seqlen if i+win>seqlen else i+win
        yield seq[i:j]
        if j==seqlen: break
    return


def findKmers(finput, ksize):
    df = pd.DataFrame()
    stepw = 1
    characters = ["G", "A", "T", "C"]
    perms(ksize, characters)
    permsList = [''.join(str(i) for i in x) for x in product(characters, repeat=ksize)]

    for seq_record in SeqIO.parse(finput, "fasta"):
        seq = seq_record.seq
        seq = seq.upper()
        name_seq = seq_record.name
        probabilities = []
        entropy_list = []
        #for k in range(1, ksize+1):
        kmer = {}
        totalWindows = (len(seq) - ksize) + 1 # (L - k + 1)
        for key in permsList:
            kmer[key] = 0
        #print(kmer)
        for subseq in chunksTwo(seq, ksize):
            if subseq in kmer:
                # print(subseq)
                kmer[subseq] = kmer[subseq] + 1
            else:
                kmer[subseq] = 1
        #print(kmer.items())
        for key, value in kmer.items():
            prob = value/totalWindows
            probabilities.append(prob)
            if prob == 0:
              entropy_equation = 0
            else:
              entropy_equation = prob * math.log(prob, 2)
            entropy_list.append(entropy_equation)
        entropy = -(sum(entropy_list))
        ext_feat = probabilities
        ext_feat.append(entropy)
        df[name_seq] = ext_feat
    #print(df)
    # transpose the dataframe to use describe to calculate mean and std
    df2 = df.T
    df3 = df2.describe().loc[['mean', 'std', 'max']]
    df3 = df3.unstack().to_frame().T
    return df3


#df = findKmers("./proc/preproc/affected/SRS066334_pp.fsa",3)#############################################################################
#print(df)
```

```python
# adaptation of the MathFeature package

import argparse
import numpy as np
from Bio import SeqIO
from itertools import product
from scipy.fftpack import fft, ifft
import warnings
import sys
import scipy.stats
import statistics
import os
import collections
warnings.filterwarnings('ignore')


#############################################################################
#############################################################################

def sequence_length():
    dlength = []
    for name, label in dataset_labels.items():
        length = 0
        for seq_record in SeqIO.parse(name, 'fasta'):
            seq = seq_record.seq
            if len(seq) > length:
                length = len(seq)
        dlength.append(length)
    # max_length = max(dlength)
    return max(dlength)

#############################################################################
#############################################################################


def header_fourier():
    dataset = open(foutput, 'a')
    dataset.write('nameseq,average,median,maximum,minimum,peak,'
                  + 'none_levated_peak,sample_standard_deviation,population_standard_deviation,'
                  + 'percentile15,percentile25,percentile50,percentile75,amplitude,'
                  + 'variance,interquartile_range,semi_interquartile_range,'
                  + 'coefficient_of_variation,skewness,kurtosis,label')
    dataset.write('\n')
    return



def feature_extraction(features, spectrum, spectrumTwo):
    average = sum(spectrum)/len(spectrum)
    features.append(average)
    ###################################
    median = np.median(spectrum)
    features.append(median)
	###################################
    maximum = np.max(spectrum)
    features.append(maximum)
    ###################################
    minimum = np.min(spectrum)
    features.append(minimum)
    ###################################
    peak = (len(spectrum)/3)/(average)
    features.append(peak)
    ###################################
    peak_two = (len(spectrumTwo)/3)/(np.mean(spectrumTwo))
    features.append(peak_two)
    ###################################
    standard_deviation = np.std(spectrum) # standard deviation
    features.append(standard_deviation)
    ###################################
    standard_deviation_pop = statistics.stdev(spectrum) # population sample standard deviation
    features.append(standard_deviation_pop)
    ###################################
    percentile15 = np.percentile(spectrum, 15)
    features.append(percentile15)
    ###################################
    percentile25 = np.percentile(spectrum, 25)
    features.append(percentile25)
    ###################################
    percentile50 = np.percentile(spectrum, 50)
    features.append(percentile50)
    ###################################
    percentile75 = np.percentile(spectrum, 75)
    features.append(percentile75)
    ###################################
    amplitude = maximum - minimum
    features.append(amplitude)
    ###################################
    # mode = statistics.mode(spectrum)
    ###################################
    variance = statistics.variance(spectrum)
    features.append(variance)
    ###################################
    interquartile_range = np.percentile(spectrum, 75) - np.percentile(spectrum, 25)
    features.append(interquartile_range)
    ###################################
    semi_interquartile_range = (np.percentile(spectrum, 75) - np.percentile(spectrum, 25))/2
    features.append(semi_interquartile_range)
    ###################################
    coefficient_of_variation = standard_deviation/average
    features.append(coefficient_of_variation)
    ###################################
    skewness = (3 * (average - median))/standard_deviation
    features.append(skewness)
    ###################################
    kurtosis = (np.percentile(spectrum, 75) - np.percentile(spectrum, 25)) / (2 * (np.percentile(spectrum, 90) - np.percentile(spectrum, 10)))
    features.append(kurtosis)
    ###################################
    return features


# return data frame of the features
def accumulated_nucle_frequency_fourier(finput):
    df = pd.DataFrame()
    for seq_record in SeqIO.parse(finput, 'fasta'):
        seq = seq_record.seq
        seq = seq.upper()
        name_seq = seq_record.name
        features = []
        spectrum = []
        spectrumTwo = []
        mapping = []
        A = 0
        C = 0
        T = 0
        G = 0
        for i in range(len(seq)):
            if seq[i] == 'A':
                A += 1
                mapping.append(A / (i + 1))
            elif seq[i] == 'C':
                C += 1
                mapping.append(C / (i + 1))
            elif seq[i] == 'T' or seq[i] == 'U':
                T += 1
                mapping.append(T / (i + 1))
            else:
                G += 1
                mapping.append(G / (i + 1))
        Fmap = fft(mapping)
        for i in range(len(mapping)):
            specTotal = (abs(Fmap[i])**2)
            specTwo = (abs(Fmap[i]))
            spectrum.append(specTotal)
            spectrumTwo.append(specTwo)
        features = feature_extraction(features, spectrum, spectrumTwo)
        df[name_seq] = features
    #print(df)
    # transpose the dataframe to use describe to calculate mean and std
    df2 = df.T
    df3 = df2.describe().loc[['mean']]
    df3 = df3.unstack().to_frame().T
    return df3

#df = accumulated_nucle_frequency_fourier("./proc/preproc/affected/SRS066334_pp.fsa")#############################################################################
#print(df)
```

```python
from Bio import SeqIO

def fasta_lengths_hist(finput):
  bins = 40
  lengths = [len(record.seq) for record in SeqIO.parse(open(finput), 'fasta')]
  hist = np.histogram(np.array(lengths), bins)
  df = pd.DataFrame(hist[0])
  return df.T

#df = fasta_lengths_hist("./proc/preproc/affected/SRS066334_pp.fsa")#############################################################################
#df
```

```python
# using the renamer class code represented in: https://stackoverflow.com/questions/40774787/renaming-columns-in-a-pandas-dataframe-with-duplicate-column-names

class renamer():
  def __init__(self):
    self.d = dict()

  def __call__(self, x):
      if x not in self.d:
          self.d[x] = 0
          return x
      else:
          self.d[x] += 1
          return "%s_%d" % (x, self.d[x])
```

```python
# show how the k-mer frequency is looking for all the files

import matplotlib.pyplot as plt
import seaborn as sns

def plot_k_mers(df, finputs):
  num_files = len(finputs)
  # creating the subplot - 2 columns of histograms
  fig, axes = plt.subplots(2,int(num_files/2), figsize=(40,20))
  fig.tight_layout(pad=2.5)
  for i in range(len(finputs)):
    ax_cur = axes[int(i%2)][int(i/2)]
    file = finputs[i]
    label = file.split('/')[-2]
    sample_name = file.split('/')[-1].split('.')[0]
    df_sample = df[df["sample"]==sample_name]
    df_sample = df_sample.iloc[:,61:]
    df_sample = df_sample.T
    # the data frame is built as mean column and then std - so we would like to take only the mean value
    df_sample = df_sample.reset_index()
    df_sample = df_sample.iloc[lambda x: x.index % 2 == 1]
    df_sample['k-mer'] = range(len(df_sample))
    df_sample = df_sample.iloc[:,1:]
    data_col_name = list(df_sample.columns)[0]
    df_sample = df_sample.rename(columns={data_col_name:"frequency","k-mer":"k-mer"})
    g = sns.lineplot(data=df_sample, x="k-mer", y="frequency", ax=ax_cur, marker='o')
    g.set_title(label)
```

# Building a classifier for the disease

```python
from sklearn.model_selection import train_test_split

vector = df.iloc[:,1:]
col_names = [f"Feature {i}" for i in range(len(list(vector.columns))-1)]
print(col_names)
col_names.insert(0,"label")
vector.columns = col_names
```

```python
!pip install catboost
```

```python
from catboost import CatBoostClassifier

vector_aff = vector[vector["label"]==1]
vector_naff = vector[vector["label"]==0]
train_aff, test_aff = train_test_split(vector_aff, test_size=0.33)
train_naff, test_naff = train_test_split(vector_naff, test_size=0.33)
train = pd.concat([train_aff, train_naff])
test = pd.concat([test_aff, test_naff])
X_train = train.drop(['label'], axis=1)
X_test = test.drop(['label'], axis=1)
Y_train = train["label"].values
Y_test = test["label"].values

model = CatBoostClassifier(iterations=100,
                          learning_rate=0.1,
                          depth=15)
model.fit(X_train, Y_train)
```

```python
from sklearn.metrics import accuracy_score

y_predict = model.predict(X_test)
accuracy_score(Y_test,y_predict)
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

plt.rcParams.update({'font.size': 40})
cm = confusion_matrix(Y_test, y_predict)
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
fig, ax = plt.subplots(figsize=(40,40))
ax.grid(False)
disp.plot(ax=ax)
```

```python
importances = model.get_feature_importance(type='PredictionValuesChange')
feature_importances = pd.Series(importances, index=X_test.columns).sort_values()
print(feature_importances)
# the most important features in the model
print(feature_importances.tail(10))
```
